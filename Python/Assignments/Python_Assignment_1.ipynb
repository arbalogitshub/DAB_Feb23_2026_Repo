### Task 1: Variable Declaration and Data Types
* Declare variables of different types (integer, float, string, boolean, etc.).
* Print their values and types using the type() function.

```python
# Declare integer variable and print its value and its type.
my_integer = 8 
print(f"my_integer: {my_integer}, {type(my_integer)}\n")

# Declare float variable and  print its value and its type.
my_float = 4.5 
print(f"my_float: {my_float}, {type(my_float)}\n")

# Declare string variable and print its value and its type.
my_string = "Boris Karloff" 
print(f"my_string: {my_string}, {type(my_string)}\n")

# Declare boolean variable and print its value and its type.
my_bool= True
print(f"my_bool: {my_bool}, {type(my_bool)}")
```

## Task 2: String Manipulation
* Create a string variable.
* Apply at least three different string methods to modify or retrieve information from the string.

```python
# Create string variable and print variable.
# This has whitespace on left side of string.
# Strings are immutable objects.
snake_message = " Python is cool" 

# Output -> " Python is cool" 
print(f"snake_message original value: {snake_message}\n")

# Modification 1: Prints out value of snake_message in all lowercase using lower(). 
print(f"snake_message value when using lower(): {snake_message.lower()}\n")

# Modification 2: Prints out value of snake_message and removes whitespaces on left side of the string value.
print(f"snake_message value when using lstrip(): {snake_message.lstrip()}\n")

# Modification 3: Prints out value of snake_message and swaps cases in the string.
print(f"snake_message value when using swapcase(): {snake_message.swapcase()}\n")

# Python returns a new string, but snake_message still keeps the original value because we did not reassign the new string to our variable.
print(f"We still have the original value of snake_message: {snake_message}\n")
```

## Task 3: Basic Operations
Perform and print the results of the following operations:
* Arithmetic operations (+, -, *, /, %, **)
* Comparison operations (>, <, ==, !=, >=, <=)
* Logical operations (and, or, not)

```python
# Arithmetic operations 
a,b = 5,6 
print(f"a is {a}, and b is {b}\n")
print("Performing Arithmetic Operations:\n")
print(f"Addition: {a+b}")
print(f"Subtraction: {a-b}")
print(f"Multiplication: {a*b}")
# For Division: This will give us a decimal which will be be a float data type.
print(f"Division: {a/b}") 
print(f"Exponentiation: {a**b}")
# For Remainder below: % calculates the remainder of a divided by b.
print(f"Remainder: {a%b}") 
```

```python
# Comparison operations (>, <, ==, !=, >=, <=)
a,b,c,d = 32,-32,57,60
print(f"a is {a}, b is {b}, c is {c} and d is {d}\n")

print(f"{a} is greater than {b}: {a > b}")
print(f"{a} is less than {d}: {a < d}")
print(f"{a} is not equal to {b}: {a != b}")
print(f"{d} is less than or equal to {d}: {d <= d}")
print(f"{c} is greater than or equal to {b}: {c >=  b}")
print(f"{b} is equal {a*-1}: {b == a*-1 }")
```

## Task 4: Data Collections
Create examples of the following data structures. Access and print specific elements from each. :
* List
* Dictionary
* Tuple<br>


```python
# Create vegetables list and print entire list.
vegetables = ['carrots', 'potatoes', 'radishes', 'cauliflower', 'broccoli']
print(f"This is my entire vegetables list: {vegetables}\n") 

# Print vegetables from index 2 through index 3 (4th index is excluded)
print(f"This is a subset of vegetables list: {vegetables[2:4]}\n") 

# Display the last item in the list 
print(f"This is the last vegetable in my vegetables list: {vegetables[-1]}") 
```

```python
# Create a dictionary.
drink_order = {"item":"coffee", "price":1.99 , "calories": 5, "temperature": "hot"}
print(f"This is all the keys and values in drink_order dictionary: {drink_order}\n")

# Displays the values of these keys: "item", and "calories" from our drink_order dictionary
print(f"This {drink_order['item']} has {drink_order['calories']} calories")
```

```python
# Create a tuple. 
apes_tuple = ("gorillas", "orangutans", "chimpanzees", "humans")
print(f"These are all the apes in apes_tuple: {apes_tuple}\n")

# Displays apes in index 1 through index 2 (excludes index 3).
print(f"These are my favorite apes: {apes_tuple[1:3]}")
```

## Task 5: Expression Combination
* Write a Python expression that combines multiple variables and operations into a single meaningful expression.
* Print the result.

```python
active = True

user_amount = 5_000
insert_amount = 500 
request_amount = 4_000

# Simulate an ATM Machine. 
# We will only have 1 input option for now. 
# Very Very Very basic ATM

while active == True:
    print("Welcome to this ATM!\n") \
    # We will only use this input() once 
    input_letter = input(f"\nWhat service would you like today? \n a. Display Amount \n b. Insert Money \n c. Take Money\nType in a, b, c or another letter to quit:")
    
    print("\n")
    
    if (input_letter == "a"):
        print(f"You have {user_amount} in your account") 
        print("Goodbye!") 
        
    elif(input_letter == "b"):
        
        if(insert_amount >= 0):
             print(f"You have inserted {insert_amount} in your account")
             print(f"You now have {user_amount + insert_amount} in your account\n")
            
    elif (input_letter == "c"):
        
        if (request_amount == user_amount or request_amount > user_amount):
            print(f"You have taken all of your money in your account")
            print(f"You now have no money in your account left\n")
        else:
             print(f"You have requested to take {request_amount} in your account")
             print(f"You now have {user_amount - request_amount} in your account\n")
    else: 
        
        print("Goodbye!") 
        active = False
```
