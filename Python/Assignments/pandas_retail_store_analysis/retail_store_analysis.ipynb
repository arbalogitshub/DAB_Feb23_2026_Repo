## PART 2: Data Cleaning
* Handling Missing Values
* Removing Duplicates


### Data Cleaning in Customers DataFrame

```python
# Identify Missing Data in Customers
print("====== Customers Missing Values =====") 
print(customers.isnull().sum())


print('\n')

# Review Data Types of the dataframes
print("====== Customers Data Types =====") 
print(customers.dtypes)
print('\n')

# Replace missing data in customers zip_code 
if 'zip_code' in customers.columns:

    print("====== Zip Code Before Cleaning Missing Values =====") 
    missing_zip_cnt_b4 = customers['zip_code'].isna().sum()
    print(f"Number of Missing Values in zip_code column: {missing_zip_cnt_b4}\n")
    
    print("====== Zip Code After Cleaning Missing Values =====") 
    # Using mode()[0] replaces all missing values in zip code with single value 
    customers['zip_code'] = customers['zip_code'].fillna(customers['zip_code'].mode()[0])
    missing_zip_cnt_after = customers['zip_code'].isna().sum()
    print(f"Number of Missing Values in zip_code column: {missing_zip_cnt_after}")

# Replace Missing Data in customers phone
if 'phone' in customers.columns:

    print("\n====== Phone Before Cleaning Missing Values =====") 
    missing_phone_cnt_b4 = customers['phone'].isna().sum()
    print(f"Number of Missing Values in phone column: {missing_phone_cnt_b4}\n")
    
    print("====== Phone After Cleaning Missing Values =====")  
    customers['phone'] = customers['phone'].fillna('000-000-0000')
    missing_phone_cnt_after = customers['phone'].isna().sum()
    print(f"Number of Missing Values in phone column: {missing_phone_cnt_after}")

# Replace Missing Data in customers email
if 'email' in customers.columns:

    print("\n====== Email Before Cleaning Missing Values =====") 
    missing_email_cnt_b4 = customers['email'].isna().sum()
    print(f"Number of Missing Values in email column: {missing_email_cnt_b4}\n")
    
    print("====== Email After Cleaning Missing Values =====")  
    customers['email'] = customers['email'].fillna('Unknown_Email')
    missing_email_cnt_after = customers['phone'].isna().sum()
    print(f"Number of Missing Values in email column: {missing_email_cnt_after}")

# Replace Missing Data in customers registration_date
if 'registration_date' in customers.columns:

    print("\n====== Registration Date Before Cleaning Missing Values =====") 
    missing_rd_cnt_b4 = customers['registration_date'].isna().sum()
    print(f"Number of Missing Values in registration_date column: {missing_rd_cnt_b4}\n")
    
    print("====== Registration Date After Cleaning Missing Values =====")  
    customers['registration_date'] = customers['registration_date'].bfill()
    missing_rd_cnt_after = customers['registration_date'].isna().sum()
    print(f"Number of Missing Values in registration_date column: {missing_rd_cnt_after}")
```

### Data Cleaning in Products DataFrame

```python
# Identify Missing Data in Products
print("====== Products Missing Values =====") 
print(products.isnull().sum())
print('\n')

# Review Data Types of the dataframes
print("====== Products Data Types =====") 
print(products.dtypes)
print('\n')

if 'weight' in products.columns:
    print("====== Weight Before Cleaning Missing Values =====") 
    products['weight'].isna().sum()
    print(f"Number of Missing Values in weight column: {products['weight'].isna().sum()}\n")
    
    print("====== Weight After Cleaning Missing Values =====")  
    # df['age']= df['age'].fillna(df['age'].median())
    products['weight'] = products['weight'].fillna(products['weight'].mean())
    print(f"Number of Missing Values in weight column: {products['weight'].isna().sum()}")
```

### Data Cleaning in Stores DataFrame

```python
# Idenitfy missing Data in Stores DataFrame
print("====== Stores Missing Values =====") 
print(stores.isnull().sum())
print('\n')

# Review Data Types of the dataframes
print("====== Stores Data Types =====") 
print(stores.dtypes)
print('\n')

if 'size_sqft' in stores.columns:
    print("====== Size_Sqft After Cleaning Missing Values =====")  
    stores['size_sqft'] = stores['size_sqft'].fillna(stores['size_sqft'].mean())
    print(f"Number of Missing Values in size_sqft column: {stores['size_sqft'].isna().sum()}\n")

if 'opening_date' in stores.columns:
    print("====== Opening Dates After Cleaning Missing Values =====")  
    stores['opening_date'] = stores['opening_date'].bfill()
    print(f"Number of Missing Values in opening_date column: {stores['opening_date'].isna().sum()}\n")
```

### Data Cleanning in Sales DataFrame

```python
# Identify missing data in Sales
print("====== Sales Missing Values =====") 
print(sales.isnull().sum())
print('\n')

# View Sales data types
print("====== Sales Data Types =====") 
print(sales.dtypes)
print('\n')

if 'date' in sales.columns:
    print("====== Date After Cleaning Missing Values =====")  
    sales['date'] = sales['date'].bfill()
    print(f"Number of Missing Values in date column: {sales['date'].isna().sum()}\n")

if 'payment_method' in sales.columns:
    print("====== Payment Method After Cleaning Missing Values =====")  
    sales['payment_method'] = sales['payment_method'].fillna(sales['payment_method'].mode()[0])
    print(f"Number of Missing Values in payment_method column: {sales['payment_method'].isna().sum()}\n")

if 'store_id' in sales.columns:
    print("====== Store Id  After Cleaning Missing Values =====")  
    sales['store_id'] = sales['store_id'].fillna(sales['store_id'].mean())
    print(f"Number of Missing Values in store_id column: {sales['store_id'].isna().sum()}\n")
```

### Data Cleaning in Inventory DataFrame

```python
print("====== Inventory Missing Values =====") 
print(inventory.isnull().sum())
print('\n')

# View Inventory Data Types
print("====== Inventory Data Types =====") 
print(inventory.dtypes)
print('\n')

if 'last_restock_date' in inventory.columns:
    print("====== Last Restock Date After Cleaning Missing Values =====")  
    inventory['last_restock_date'] = inventory['last_restock_date'].bfill()
    print(f"Number of Missing Values in last_restock_date column: {inventory['last_restock_date'].isna().sum()}\n")

if 'reorder_level' in inventory.columns:
    print("====== Reorder Level After Cleaning Missing Values =====")  
    inventory['reorder_level'] = inventory['reorder_level'].fillna(inventory['reorder_level'].mean())
    print(f"Number of Missing Values in reorder_level column: {inventory['reorder_level'].isna().sum()}\n")
```
