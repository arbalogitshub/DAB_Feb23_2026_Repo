### Load all three CSV files and convert date columns to proper DateTime format using pd.to_datetime()

def load_data_and_review(data_path) is a function that imnports the data and shows the following information: 
* Checks if file can be imported.
* Preview of Data (First 5 rows displayed).
* Calls columns_to_datetime to convert date columns to DateTime format.
* Obtains DataFrame information with data.info()
* Obtains missing value & duplicate rows count .

def columns_to_datetime(data) is a function that takes the imported data, and converts the data columns to the proper DateTime() format:
* Checks if DataFrame is empty
* Searches for columns where 'date' is part of column name
* Checks if found columns already have datetime format and if not in format then makes conversion.
* Reports conversion issues
* if no date columns found then reports back 

We will use these functions for each dataframe 


```python
import pandas as pd
from typing import Optional

# Loads the data and displays information about imported data 
def load_data_and_review(data_path):
    print("""Load CSV and perform initial data review.\n""")
   
    try:
        data = pd.read_csv(data_path)
        print(f"Successfully loaded {data_path}\n")
    except FileNotFoundError:
        print(f"Error: File '{data_path}' not found\n")
        raise
    except Exception as e:
        print(f"Error loading file: {e}\n")
        raise

    data = columns_to_datetime(data)

     # Preview
    print("=== Preview Data ===\n")
    print(data.head())
    
    # Info
    print("=== DataFrame Info ===\n")
    data.info()
    
    # Missing values summary
    missing_data = data.isnull().sum()
    if missing_data.sum() > 0:
        print("=== Missing Values (columns with >0 missing) ===\n")
        print(missing_data[missing_data > 0])
    
    # Duplicates
    dup_count = data.duplicated().sum()
    print(f"\n=== Duplicate Rows: {dup_count} ===\n")
    
    return data
    
# Converts dataframe columns to datetime format 
def columns_to_datetime(data: pd.DataFrame) -> pd.DataFrame:
    if data.empty:
        print("Warning: Empty DataFrame provided\n")
        return data
    
    date_columns_found = False
    
    for col in data.columns:
        if 'date' in col.lower():
            date_columns_found = True
            original_dtype = data[col].dtype
            
            # Skip if already datetime
            if pd.api.types.is_datetime64_any_dtype(data[col]):
                print(f"=== {col} already datetime type, skipping ===\n")
                continue
                
            print(f"=== {col} - Original type: {original_dtype} ===\n")
            
            # Convert with error handling
            data[col] = pd.to_datetime(data[col], errors='coerce')
            
            # Report conversion issues
            if data[col].isna().sum() > 0:
                print(f"Warning: {data[col].isna().sum()} values couldn't be converted in column '{col}\n'")
    
    if not date_columns_found:
        print("No date columns found (looking for 'date' in column names)\n")
    
    return data
```

### Extract useful components from the sales transaction dates: year, month, and day of week
### Display the date range covered by your sales data (earliest and latest dates)

```python
# Extract useful components from the sales transaction dates: year, month, and day of week
sales['td_year'] = sales['transaction_date'].dt.year #grab year from date
sales['td_month'] = sales['transaction_date'].dt.month #grab month from date
sales['td_day_of_week'] = sales['transaction_date'].dt.day_name() #grab day of the week from date

# Display sample data of sales transaction_date and the newly made date columns extracted from sales transaction_date
print("--- Preview of Sales Extracted Transaction Dates---")
print(sales[['transaction_date','td_year', 'td_month', 'td_day_of_week']].head())

# Display the data range covered by earliest and latest dates 
earliest_date= sales['transaction_date'].min() # Earliest date in Sales data 
latest_date= sales['transaction_date'].max() # Latest date in Sales data 

print(f"\n--- Sales Date Range ---")
print(f"Earliest Date: {earliest_date}")
print(f"Latest Date: {latest_date}")
```

### Basic Time Analysis:
- Find the busiest day of the week for sales (by counting transactions)
- Determine which month had the most transactions
- Count how many sales happened in each year

### Find the businest day of the week for sales 

```python
# Busiest Day of the Week 
weekday_transactions = sales.groupby('td_day_of_week')['sale_id'].count() # Get Count of sale_ids grouped by td_day_of_week

# Obtain the name of the day with max count 
busiest_day = weekday_transactions.idxmax()

# Get corresponding maxcount value 
max_count = weekday_transactions.max()

print("--- Total Transactions per Weekday ---")
print(weekday_transactions)

print("\n--- Busiest Weekday by Transaction Count ---")

print (f"Busiest day of the Week: {busiest_day}")
print(f"Transaction Counts: {max_count}")
```

### Determine which month had the most transactions

```python
# Busiest Day of the Week 
monthly_transactions = sales.groupby('td_month')['sale_id'].count() # Get Count of sale_ids grouped by td_month

# Obtain the name of the month with max count 
busiest_month= monthly_transactions.idxmax()

# Get corresponding maxcount value 
max_count = monthly_transactions.max()

print("--- Total Transactions per Month ---")
print(monthly_transactions)

print("\n--- Busiest Month by Transaction Count ---")

print (f"Busiest Month: {busiest_month}")
print(f"Transaction Counts: {max_count}")
```

### Count how many sales happened in each year


```python
sales_count_per_year = sales.groupby('td_year')['sale_id'].count().sort_values(ascending = False) # Get Count of sale_ids grouped by td_year

print("--- Total Transactions per Year ---")
print(sales_count_per_year)
```

## Part 2: GroupBy and Aggregation

### Product Analysis:
* Group products by category and calculate the average price for each category
* Count how many products are in each category

```python
# product_analysis displays number of products per category, and 
# displays average price of price per category.

product_analysis = products.groupby('category').agg(
    products_in_category = ('product_id', 'count'),
    average_category_price = ('price', 'mean')
).round(2).sort_values('average_category_price', ascending=False)

print(product_analysis)
```

### Customer Analysis:
- Group customers by state and find the average age in each state
- Calculate total spending per customer (sum up all their purchases)
- Find customers who have made the most purchases

### Group customers by state and find the average age in each state


```python
# Get count of customers per state, and average age of customer per state.
state_analysis = customers.groupby('state').agg(
        customer_count = ('customer_id', 'count'),
        average_age = ('age', 'mean')
).round(2)

print("--- Customer State Analysis ---")
print(state_analysis)
```

### Calculate total spending per customer (sum up all their purchases)
### Find customers who have made the most purchases


```python
# Get total spending amount per customer and number of transactions per customer
# FYI we sorted this by total amount spent per customer.
customers_analysis = sales.groupby('customer_id').agg(
        total_spending = ('total_amount', 'sum'),
        total_transactions = ('sale_id', 'count')
).sort_values(by ='total_spending', ascending = False)

print("\n--- Customer Sales Analysis ---")
print(customers_analysis)

# obtain the highest count of purchases from customer_analysis
max_purchases = customers_analysis['total_transactions'].max()

# Get all customers with the most purchases 
top_customers = customers_analysis[customers_analysis['total_transactions'] == max_purchases]

print(f"\n--- Top Customers ({len(top_customers)} customers with {max_purchases} purchases each) ---")
print(top_customers)

```

### Sales Analysis 
* Calculate total sales revenue by month
* Find the average transaction amount by day of the week
* Determine which product category generates the most total revenue

### Calculate total sales revenue by month

```python
sales_by_month = sales.groupby('td_month')['total_amount'].sum()

print("-- Total Sales Revenue by Month ---")
print(sales_by_month)
```

 ### Find the average transaction amount by day of the week


```python
# Define the correct day order
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Convert to categorical with specified order
sales['td_day_of_week'] = pd.Categorical(
    sales['td_day_of_week'], 
    categories=day_order, 
)

# Average Transactions by Day of the Week
average_transactions_dow = sales.groupby('td_day_of_week', observed=True)['total_amount'].mean().round(2)

print("-- Average Transaction Amount by Day of the week ---")
print(average_transactions_dow)
```

### Determine which product category generates the most total revenue

```python
# Use Merge to combine products and sales dataframes together
merged = pd.merge(products,sales, on = 'product_id')

# Filtered only the necessary columns to answer the question
result = merged[['category','product_id','sale_id','total_amount']]

# Preview filtered merged dataframes
print("--- Preview of Merged Products & Sales Data ---\n")
print(result.head())
print('\n')

# Obtain total revenue per category
total_revenue_category = result.groupby('category')['total_amount'].sum().sort_values(ascending=False)

# Display Revenue 
print("--- Total Revenue per product Category---")
print(total_revenue_category)

# Otain max revenue per category amount and category name with max amount.
max_revenue_amount = total_revenue_category.max()
max_revenue_name = total_revenue_category.idxmax()


# Display Maximum Revenue 
print("\n--- Category that Generated the Most Revenue ---")
print(f"Product Category: {max_revenue_name}\nMax Revenue: ${max_revenue_amount}")
```
