### Task 5: Function that Returns a Value
Write a function named get_full_name that combines first, middle, and last names.

* The middle name should be optional.
* Return the full name as a string.

```python
def get_full_name (first_name, last_name, middle_name =""):
    full_name = ""

    # Fix names: Remove whitespaces in names 
    first_name = first_name.replace(" ", "").replace("\n", "").replace("\t", "")
    middle_name = middle_name.replace(" ", "").replace("\n", "").replace("\t", "")
    last_name = last_name.replace(" ", "").replace("\n", "").replace("\t", "")

    # Fix names: Make it into a title format for each name 
    first_name = first_name.title()
    middle_name = middle_name.title()
    last_name = last_name.title()
    
    # If middle name is default, replace with a space
    if middle_name == "":
        full_name = first_name + ' ' + last_name
        return full_name 
    else:
        full_name = first_name + ' '+ middle_name + ' '+ last_name 
        return full_name

# Test Cases. 
print(get_full_name('Ma ri   a','O r t   i \tz','guaDALUPE'))
print(get_full_name('Gustavo','Guzman','Cornelious'))
print(get_full_name('DIO','Brando'))
print(get_full_name('Johnny','Bravo',''))
```

### Task 6: Function Composition

* C = (F - 32) * 5/9

```python
# Converts C to F.
def celsius_to_fahrenheit(convert_celsius):
    fahrenheit = (convert_celsius * 9/5) + 32
    print (f"{convert_celsius} Celsius = {fahrenheit} Farenheit!") 

# Converts F to C.
def fahrenheit_to_celsius(convert_farenheit):
    celsius = (convert_farenheit - 32) * 5/9
    celsius=round(celsius,1)
    print (f"{convert_farenheit} Farenheit = {celsius} Celsius!") 

# Test Cases.
celsius_to_fahrenheit(0)
celsius_to_fahrenheit(15)
celsius_to_fahrenheit(25)
print("\n") 
fahrenheit_to_celsius(32)
fahrenheit_to_celsius(59)
fahrenheit_to_celsius(77)
```

### Task 7: Combining Functions and Conditionals
Write a function named analyze_temperature that:

* Takes a temperature value and a unit ('C' or 'F').
* Converts the temperature to the other unit.
* Returns a description of the weather:
    * Below 0°C: "freezing"
    * 0-15°C: "cold"
    * 15-25°C: "moderate"
    * Above 25°C: "hot"

```python

def celsius_to_fahrenheit(temp_value):
    fahrenheit = (temp_value * 9/5) + 32
    fahrenheit= round(fahrenheit,1)
    
    return fahrenheit

def fahrenheit_to_celsius(temp_value):
    celsius = (temp_value - 32) * 5/9
    celsius = round(celsius,1)
    
    return celsius

def analyze_temperature(temp_value, unit):
    if (unit.upper() == 'F'):
        value = fahrenheit_to_celsius(temp_value)
        
        if value < 0:
            return "freezing"
        elif value <= 15:
            return "cold"
        elif value <= 25:
            return "moderate"
        else:
            return "hot"
        
    elif(unit.upper() == 'C'):
        value =celsius_to_fahrenheit(temp_value)
        if value < 32:
            return "freezing"
        elif value <= 59:
            return "cold"
        elif value <= 77:
            return "moderate"
        else:
            return "hot"
    else:
        print(f"Value_Error: {unit} is not accepted!")
# Test Cases 
print(analyze_temperature (100, 'f'))
print(analyze_temperature (25, 'c'))
analyze_temperature (32, 'Q')
```

### Task 8: Function with Variable Arguments
Write a function named calculate_average that:

* Accepts any number of numeric arguments.
* Returns the average of all numbers provided.

```python
def calculate_average (*numeric_arguements):
    # Create an empty list of numbers
    numbers = []

    # Create For Loop that goes through each arugement to check if it's a numeric value
    for numeric_arg in numeric_arguements:
        if isinstance(numeric_arg, (int, float)):
            numbers.append(numeric_arg)
        else:
            # Skip if the arguement is not numeric
            None
    return round(sum(numbers)/len(numbers),2)
   

print(calculate_average (1,2,3,4,5,6,7,8))
print(calculate_average (1,2,3,4,5,6))
print(calculate_average (1,2,3))
```

### Task 9: Advanced Challenge - Student Grade Function
Write a function named calculate_final_grade that:

1) Takes a student's name and any number of assignment scores.
2) Calculates the average score.
3) Determines the letter grade:
    * 90+: A
    * 80-89: B
    * 70-79: C
    * 60-69: D
    * Below 60: F
4) Returns a formatted string:
    * Example: "Alice's final grade is 85 (B)."


```python
def calculate_final_grade(student_name, *student_scores_arguements):

    # *arguement allows us to put in any number of arguemnts 
    
    # Use this empty list to store scores in.
    scores = []

    # We add the score arguements into an empty list.
    for student_scores_arguement in student_scores_arguements:
        if isinstance(student_scores_arguement, (int, float)):
            scores.append(student_scores_arguement)

    # Get the average, to determine the final score
    final_score = round(sum(scores)/len(scores),2)
    
    full_string = ""
    if final_score >= 90:
        return f"{student_name}'s final grade is {final_score} (A)."
    elif final_score >= 80:
        return f"{student_name}'s final grade is {final_score} (B)."
    elif final_score >= 70:
        return f"{student_name}'s final grade is {final_score} (C)."
    elif final_score >= 60:
        return f"{student_name}'s final grade is {final_score} (D)."
    else: 
        return f"{student_name}'s final grade is {final_score} (F)."
    
    
# Test Cases 
print(calculate_final_grade('Thor',89,85,100,99))     # Get an A 
print(calculate_final_grade('Thor',69,60,100,100))    # Get a B 
print(calculate_final_grade('Thor',60,100,70,70))     # Get a C 
print(calculate_final_grade('Thor',69,60,62,60,61,60))# Get a D 
print(calculate_final_grade('Thor',89,0,0))           # Get a F 
```
