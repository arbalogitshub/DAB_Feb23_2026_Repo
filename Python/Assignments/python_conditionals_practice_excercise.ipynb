### Task 1: Basic if Statement
Write a function that checks if a number is positive and prints an appropriate message.

If the number is greater than 0, print "The number is positive"
Example Call:

check_positive(5)  # Output: The number is positive

```python
# Checks number to see if it's positive.
def check_number(my_number):
    print(f"The number is positive") if my_number > 0 else print(f"The number is not positive")

# Test Cases 
check_number(0)
check_number(78)
check_number(-7)
```

### Task 2: if-else Statement
Write a function that determines if a student passed or failed based on their score.

A score of 60 or higher is a pass; otherwise, it's a fail.

```python
def passed_class_check(score):
    print(f"{score} means you pass!") if score >= 60 else print(f"{score} means you fail!")

# Test Cases
passed_class_check(60)
passed_class_check(100)
passed_class_check(55)
passed_class_check(45)
```

### Task 3: if-elif-else Statement
Write a function that converts a numeric grade to a letter grade:

90-100: A, 80-89: B, 70-79: C, 60-69: D, Below 60: F

```python
def get_grade(numeric_grade):
    if numeric_grade >= 90:
        return "A"
    elif numeric_grade >= 80:
         return "B"
    elif numeric_grade >= 70: 
         return "C"
    elif numeric_grade >= 60:
         return "D"
    else:
         return "F"

# Test Cases
print(f"You get an {get_grade(99)}")
print(f"You get a {get_grade(89)}")
print(f"You get a {get_grade(70)}")
print(f"You get a {get_grade(67)}")
print(f"You get a {get_grade(59)}")
```

### Task 4: Nested Conditionals
Write a function to determine user access based on:

Login status (True or False)
Admin status (True or False)
Premium membership (True or False)

```python
def is_prem(premium_status):
    return True if premium_status == True else False

def is_admin(admin_status):
    return True if admin_status == True else False 
    
def login_status_function(login_status,admin_status,premium_status):
    if login_status == True:
        
        if is_admin(admin_status) == True and is_prem(premium_status) == False:
            print("Welcome to the website. You have Admin Access.\n")
        
        elif  is_admin(admin_status) == False and is_prem(premium_status) == True:
            print("Welcome to the website. You have Premium Access.\n")
       
        elif is_admin(admin_status) == False and is_prem(premium_status) == False:
            print("Welcome to the website. You have Basic Access.\n")
        
        else:
            print("You can't have premium and Admin Acess!\n") 
    else:
        print("You are not logged in!\n")



# Test Cases 
login_status_function(True,True,False) # Logged in as Admin 
login_status_function(True,False,True) # Logged in as Premium
login_status_function(True,False,False) # Logged in as Basic
login_status_function(False,False,True) # Not Logged in 

    
```

## Task 5: Ternary Operator
Write a function that returns "Even" or "Odd" based on whether a number is divisible by 2, using the ternary operator.

```python
def even_number_check(my_number):
    result= 'Even' if my_number % 2 == 0 else  'Odd'
    return result

# Test Cases 
print(even_number_check(23)) # Use an Odd Number.
print(even_number_check(0))  # Use an Even Number.
print(even_number_check(-8)) # Use an Even Number.
print(even_number_check(-17)) # Use an Odd Number.
```

## Task 6: Combining Conditions
Write a function to check if a year is a leap year.

A leap year is divisible by 4, except when divisible by 100 unless also divisible by 400.

```python
def leap_year_check (year):
    if year % 400 == 0:
        print("Leap Year")
    elif year % 100 == 0:
        print("Not a Leap Year")
    elif year % 4 == 0:
        print("Leap Year")
    else:
        print("Not a Leap Year")
    

leap_year_check(2000)
leap_year_check(1900)
leap_year_check(2028)
```

### Task 7: Best Practices (Refactoring)

```python
def is_electronic(item_type):
    return False if item_type != "electronics" else True

def is_premium(price):
    return False if price < 100 else True

def classify_item(item_type,price,stock_status):

    # Outcome if item is electronic, affordable and is or is not in stock.
    if is_electronic(item_type) == True and is_premium(price) == False:
        print("Affordable electronics in stock\n") if stock_status == True else print("Affordable electronics not in stock\n")
    
    # Outcome if item is electronic, premium and is or not in stock. 
    if is_electronic(item_type) == True and is_premium(price) == True:
        print("Premium electronics in stock\n") if stock_status == True else print("Premium electronics not in stock\n")

    # Outcome if item is not electronic, affordable and is or not in stock. 
    if is_electronic(item_type) == False and is_premium(price) == True:
        print("Premium non-electronics in stock\n") if stock_status == True else print("Premium non-electronics not in stock\n")

     # Outcome if item is not electronic, affordable and is or not in stock. 
    if is_electronic(item_type) == False and is_premium(price) == False:
        print("Affordable non-electronics in stock\n") if stock_status == True else print("Affordable non-electronics not in stock\n")


# Test Case Affordable Electronics.
classify_item("electronics",10,True)
classify_item("electronics",10,False)

# Test Case Premium Electronics. 
classify_item("electronics",102,True)
classify_item("electronics",100,False)

# Test Case Affordable Non-Electronics.
classify_item("Pasta",10,True)
classify_item("toilet paper ",13,False)

# Test Case Premium Non-Electronics.
classify_item("Pasta",102,True)
classify_item("toilet paper ",100,False)
```
