## Instructions (List section of this Excercise) 
* Create a shopping list with at least 5 items
* Print out the third item in the list
* Replace the first item with a different item
* Add an item to the end of the list

```python
# Function checks if you have at least 5 items in your shopping list 
def has_five_items(my_shopping_list_items):
   return True if len(my_shopping_list_items) >= 5 else False


# Create a shopping list with at least 5 items or more.
def shopping_list_function(my_shopping_list_items):
    
    # If we have at least 5 items, we can go inside this if statement
    if(has_five_items(my_shopping_list_items) == True):
        
        # Display your original shopping list
        print(f"Here's your original shopping list: {my_shopping_list_items}")

        # Create a copy of the oringal list, so when we update my_shopping_list_items 
        # We still retain the oringal shopping list items 
        original_list_cpy = my_shopping_list_items.copy()
        
        # Display the length of your shopping list.
        print(f"You have {len(my_shopping_list_items)} items in your shopping list")
        
        # Display the third item in your shopping list 
        print(f"The third item in shopping list: {my_shopping_list_items[2]}")

        # Replace the first item with a different item
        my_shopping_list_items[0] = 'bleach' # New Item replacement  
        print(f"The new first item is in shopping list: {my_shopping_list_items[0]}")

        # Add an item to the end of the list 
        my_shopping_list_items.append('yogurt')
        print(f"The last item we added to the list was: {my_shopping_list_items[-1]}")

        # Display the final shopping list: 
        print(f"The original shopping list: {original_list_cpy}\nThe new shopping list is: {my_shopping_list_items}")

    # If we don't have 5 items, we get an "error" message. 
    else:
        print(f"You have {len(my_shopping_list_items)} items in your shopping list!\nYou need to have at least 5 items!")


# A shopping list with 5 items
shopping_list_function(['milk','mayonaise','eggs','olive oil','cat litter'])

print("\n")
# A shopping list with 3 items 
shopping_list_function(['milk','mayonaise','eggs'])
```

## Instructions (Dictionary section of this Excercise) 
* Create a dictionary of 3 grocery items with their prices
* Print a message showing one item and its price
* Add a new item to the dictionary with its price
* Calculate and print the total cost of all items

```python
# Created a dictionary of 3 grocery items and their prices. 
groceries_dict = { "1L Milk": 5.00, "23oz Arizona Ice Tea": 0.99, "Bread": 3.99} 

# Print out our dictionary.
print(f"This is our grocery list: {groceries_dict}\n")

# We cannot access a dictionary directly, we can try to convert our keys and values in our dictionary
# Then try to access it to print out the message of 1 item and 1 price
grocery_keys_list = list(groceries_dict.keys())
grocery_values_list = list(groceries_dict.values())

# Print a message showing one item and its price. 
print(f"The cost of a {grocery_keys_list[1]} is ${grocery_values_list[1]}\n")

# Add a new item to the dictionary with its price.  
groceries_dict['Snickers Bar'] = 1.35

# Display the updated dictionary. 
print(f"{groceries_dict}\n")

# Calculate and print the total cost of all items.
print(f"The total cost of all items is: ${sum(groceries_dict.values())}")
```
